```python
import numpy as np
import pandas as pd

def cyclical_encode_auto_range(df, col_name):

    #计算最小值和最大值
    min_val = train_data[col_name].min()-1
    max_val = train_data[col_name].max()
    
    #所有值都相同则返回
    if min_val == max_val:
        df[col_name + '_sin'] = 0.0
        df[col_name + '_cos'] = 1.0
        return df
    
    #将[min_val, max_val] 区间视为一个周期进行归一化
    #将特征映射到[0, 1]区间
    #然后乘以2*pi得到最终的相位
    df[col_name + '_sin'] = (np.sin(
        2 * np.pi * (df[col_name] - min_val) / (max_val - min_val)
    ) + 1) / 2
    df[col_name + '_cos'] = (np.cos(
        2 * np.pi * (df[col_name] - min_val) / (max_val - min_val)
    ) + 1) / 2
    
    return df

train_data = cyclical_encode_auto_range(train_data, 'season')
train_data = cyclical_encode_auto_range(train_data, 'mnth')
train_data = cyclical_encode_auto_range(train_data, 'hr')
train_data = cyclical_encode_auto_range(train_data, 'weekday')
test_data = cyclical_encode_auto_range(test_data, 'season')
test_data = cyclical_encode_auto_range(test_data, 'mnth')
test_data = cyclical_encode_auto_range(test_data, 'hr')
test_data = cyclical_encode_auto_range(test_data, 'weekday')
train_data
```

```python
import pandas as pd

def normalize_column(df, col_name):
    
    #计算最小值和最大值
    min_val = df[col_name].min()
    max_val = df[col_name].max()

    #防止最小值和最大值相等
    if min_val == max_val:
        df[col_name] = 0.0
    else:
        #进行归一化
        df[col_name] = (df[col_name] - min_val) / (max_val - min_val)
    
    return df


train_data = normalize_column(train_data, 'weathersit')
train_data = normalize_column(train_data, 'casual')
train_data = normalize_column(train_data, 'registered')
train_data = normalize_column(train_data, 'cnt')
test_data = normalize_column(test_data, 'weathersit')
test_data = normalize_column(test_data, 'casual')
test_data = normalize_column(test_data, 'registered')
test_data = normalize_column(test_data, 'cnt')
train_data
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

def create_dataset(data_array, n_past=96, n_future=240):
    
    X_enc_list = []
    X_dec_list = []
    Y_list = []
    
    total_length = len(data_array)
    
    for i in range(total_length - n_past - n_future):
        #encoder输入
        encoder_input = data_array[i : i + n_past]  #17列全部

        #decoder输入
        decoder_input = data_array[i + n_past : i + n_past + n_future, :16]  #只取前16列

        #目标值：
        label = data_array[i + n_past : i + n_past + n_future, 16:17]  
        
        X_enc_list.append(encoder_input)
        X_dec_list.append(decoder_input)
        Y_list.append(label)
    
    X_enc = np.array(X_enc_list)  
    X_dec = np.array(X_dec_list)  
    Y = np.array(Y_list)         
    
    return X_enc, X_dec, Y
```

```python
class TimeSeriesDataset(Dataset):
    def __init__(self, data: np.ndarray, seq_length: int = 96):
        
        super().__init__()
        self.X_enc, self.X_dec, self.Y = create_dataset(data)
        
    def __len__(self):
        return len(self.X_enc)
    
    def __getitem__(self, idx):
        x_enc = self.X_enc[idx]  
        x_dec = self.X_dec[idx]  
        y = self.Y[idx]          
        
        #转为torch.Tensor
        x_enc = torch.FloatTensor(x_enc)
        x_dec = torch.FloatTensor(x_dec)
        y = torch.FloatTensor(y)
        
        return x_enc, x_dec, y
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, feature_dim, max_len=5000):
        
        super(PositionalEncoding, self).__init__()
        self.encoding = torch.zeros(max_len, feature_dim)
        positions = torch.arange(0, max_len, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, feature_dim, 2).float() * (-torch.log(torch.tensor(10000.0)) / feature_dim))

   
        self.encoding[:, 0::2] = torch.sin(positions * div_term)  
        if feature_dim % 2 == 1:
            self.encoding[:, 1::2] = torch.cos(positions * div_term[:-1])  
        else:
            self.encoding[:, 1::2] = torch.cos(positions * div_term)

        self.encoding = self.encoding.unsqueeze(0)  

    def forward(self, x):
        
        seq_len = x.size(1)
        return x + self.encoding[:, :seq_len, :].to(x.device)
```

```python
import torch
import torch.nn as nn

class MixerMLP(nn.Module):
    def __init__(self, sequence_len, feature_dim, hidden_dim):
       
        super(MixerMLP, self).__init__()

        
        self.time_norm = nn.LayerNorm(feature_dim)  # Pre-Norm
        self.time_mixing = nn.Sequential(
            nn.Linear(sequence_len, hidden_dim*2),
            nn.ReLU(),
            nn.Linear(hidden_dim*2, sequence_len)
        )

        
        self.feature_norm = nn.LayerNorm(feature_dim)  # Pre-Norm
        self.feature_mixing = nn.Sequential(
            nn.Linear(feature_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, feature_dim)
        )

    def forward(self, x):
       
        x = x + self.time_mixing(self.time_norm(x).transpose(1, 2)).transpose(1, 2)

        
        x = x + self.feature_mixing(self.feature_norm(x))

        return x


class MixerModel(nn.Module):
    def __init__(self, history_len, future_len, history_feature_dim, future_feature_dim,
                 history_hidden_dim, future_hidden_dim, num_output_layers, num_layers):
        super(MixerModel, self).__init__()

        #位置编码
        self.history_positional_encoding = PositionalEncoding(history_feature_dim, max_len=history_len)
        self.future_positional_encoding = PositionalEncoding(future_feature_dim, max_len=future_len)

        #MLP
        self.history_mixer = nn.ModuleList([
            MixerMLP(history_len, history_feature_dim, history_hidden_dim) for _ in range(num_layers)
        ])
        self.history_feature_projection = nn.Linear(history_feature_dim, history_hidden_dim)  # Map 17 -> history_hidden_dim
        self.history_time_projection = nn.Linear(history_len, future_len)  # Map P -> F

        #MLP
        self.future_mixer = nn.ModuleList([
            MixerMLP(future_len, future_feature_dim, future_hidden_dim) for _ in range(num_layers)
        ])
        self.future_feature_projection = nn.Linear(future_feature_dim, future_hidden_dim)  # Map 16 -> future_hidden_dim

        #MLP
        combined_dim = history_hidden_dim + future_hidden_dim
        self.output_mixer = nn.ModuleList([
            MixerMLP(future_len, combined_dim, combined_dim) for _ in range(num_output_layers)
        ])
        self.output_projection = nn.Linear(combined_dim, 1)

    def forward(self, history, future):
        
        #为历史记录添加位置编码
        history = self.history_positional_encoding(history)

        #历史
        for layer in self.history_mixer:
            history = layer(history)
        history = self.history_feature_projection(history)  
        history = self.history_time_projection(history.transpose(1, 2)).transpose(1, 2)  
        

        
        future = self.future_positional_encoding(future)

        
        for layer in self.future_mixer:
            future = layer(future)
        future = self.future_feature_projection(future)  
        

        
        combined = torch.cat([history, future], dim=-1)  

        
        for layer in self.output_mixer:
            combined = layer(combined)
        

        
        output = self.output_projection(combined)  

        return output
```

```python

def train_one_epoch(model, dataloader, criterion, optimizer, device):
    model.train()
    running_loss = 0.0
    running_mse = 0.0
    running_mae = 0.0
    
    for x_enc, x_dec, y in dataloader:
        x_enc = x_enc.to(device)
        x_dec = x_dec.to(device)
        y = y.to(device).squeeze(-1)
        
        outputs = model(x_enc, x_dec).squeeze(-1)
        mse_loss = criterion(outputs, y)
        mae_loss = torch.mean(torch.abs(outputs - y))
        
        optimizer.zero_grad()
        #反向传播
        mse_loss.backward()
        optimizer.step()
        
        running_loss += mse_loss.item() * x_enc.size(0)
        running_mse += mse_loss.item() * x_enc.size(0)
        running_mae += mae_loss.item() * x_enc.size(0)
    
    epoch_loss = running_loss / len(dataloader.dataset)
    epoch_mse = running_mse / len(dataloader.dataset)
    epoch_mae = running_mae / len(dataloader.dataset)
    return epoch_loss, epoch_mse, epoch_mae
```

```python
def validate_one_epoch(model, dataloader, criterion, device):
    model.eval()
    running_mse = 0.0
    running_mae = 0.0
    
    with torch.no_grad():
        for x_enc, x_dec, y in dataloader:
            x_enc = x_enc.to(device)
            x_dec = x_dec.to(device)
            y = y.to(device).squeeze(-1)

            outputs = model(x_enc, x_dec).squeeze(-1)
            mse_loss = criterion(outputs, y)
            mae_loss = torch.mean(torch.abs(outputs - y))
            
            running_mse += mse_loss.item() * x_enc.size(0)
            running_mae += mae_loss.item() * x_enc.size(0)
    
    epoch_mse = running_mse / len(dataloader.dataset)
    epoch_mae = running_mae / len(dataloader.dataset)
    return epoch_mse, epoch_mae
```

```python
#设备
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#超参数
SEQ_LENGTH = 96
INPUT_SIZE = 17

LEARNING_RATE = 1e-4
BATCH_SIZE = 32
EPOCHS = 150
#超参数设置
history_len = 96  #历史时间步长 (P)
future_len = 240   #未来时间步长 (F)
history_feature_dim = 17  #历史特征维度
future_feature_dim = 16   #未来特征维度
history_hidden_dim = 128   #历史隐藏层维度
future_hidden_dim = 128    #未来隐藏层维度
num_output_layers = 4     #输出 Mixer-MLP 的层数
num_layers = 3            #历史和未来特征处理 Mixer-MLP 的层数


```

```python
#将DataFrame转为数组
train_array = train_data.values  
test_array = test_data.values   

#构建Dataset
train_dataset = TimeSeriesDataset(train_array, seq_length=SEQ_LENGTH)
test_dataset = TimeSeriesDataset(test_array, seq_length=SEQ_LENGTH)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)
```

```python
#初始化模型
model = MixerModel(
    history_len=history_len,
    future_len=future_len,
    history_feature_dim=history_feature_dim,
    future_feature_dim=future_feature_dim,
    history_hidden_dim=history_hidden_dim,
    future_hidden_dim=future_hidden_dim,
    num_output_layers=num_output_layers,
    num_layers=num_layers
).to(device)

#损失函数MSE和优化器Adam
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

best_val_mse = float('inf')

for epoch in range(EPOCHS):
    train_loss, train_mse, train_mae = train_one_epoch(model, train_loader, criterion, optimizer, device)
    val_mse, val_mae = validate_one_epoch(model, test_loader, criterion, device)
    
    #保存在验证集上表现最好的模型
    if val_mse < best_val_mse:
        best_val_mse = val_mse
        torch.save(model.state_dict(), "240our2_best_model1.pth")
    
    print(f"Epoch [{epoch+1}/{EPOCHS}] "
          f"Train Loss: {train_loss:.6f} | Train MSE: {train_mse:.6f} | Train MAE: {train_mae:.6f} "
          f"Val MSE: {val_mse:.6f} | Val MAE: {val_mae:.6f}")
```

```python
def validate_real(model, dataloader, criterion, device, min_val, max_val):
    
    model.eval()
    running_mse = 0.0
    running_mae = 0.0
    all_outputs = []  #保存所有批次的预测值
    all_labels = []  #保存所有批次的真实值

    with torch.no_grad():
        for x_enc, x_dec, y in dataloader:
            x_enc = x_enc.to(device)
            x_dec = x_dec.to(device)
            y = y.to(device).squeeze(-1)

            outputs = model(x_enc, x_dec).squeeze(-1)

            #反归一化预测值和标签
            outputs_real = outputs * (max_val - min_val) + min_val
            batch_y_real = y * (max_val - min_val) + min_val

            #保存预测值和真实值
            all_outputs.append(outputs_real.cpu().numpy())
            all_labels.append(batch_y_real.cpu().numpy())

            #计算反归一化后的MSE和MAE
            mse_loss = criterion(outputs_real, batch_y_real)
            mae_loss = torch.mean(torch.abs(outputs_real - batch_y_real))

            #累积误差
            running_mse += mse_loss.item() * x_enc.size(0)
            running_mae += mae_loss.item() * x_enc.size(0)

    #将保存的所有批次结果拼接为一个数组
    all_outputs = np.concatenate(all_outputs, axis=0)
    all_labels = np.concatenate(all_labels, axis=0)

    #计算最终的平均误差
    epoch_mse = running_mse / len(dataloader.dataset)
    epoch_mae = running_mae / len(dataloader.dataset)

    return epoch_mse, epoch_mae, all_outputs, all_labels
```

```python
#加载最优模型
model.load_state_dict(torch.load("240our2_best_model1.pth"))
model.eval()

test_data = pd.read_csv('data/test_data.csv')

min_val = test_data['cnt'].min()
max_val = test_data['cnt'].max()

mse_test, mae_test, all_outputs1, all_labels1 = validate_real(model, test_loader, criterion, device, min_val, max_val)
print(f"Final Test MSE: {mse_test:.6f}, MAE: {mae_test:.6f}")
```
